```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('airfoils/tkv2008.dat','r') as f:
    airfoil = f.read().strip()
airfoil = airfoil.split('\n')[2:]
airfoil = [row.split('\t') for row in airfoil]
airfoil = np.array(airfoil,dtype=np.float64)
```

```python
bordo_de_ataque = np.argmin(airfoil[:,0])
bordo_de_fuga   = 0
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.plot(airfoil[:,0],airfoil[:,1],'.')
ax.set_aspect('equal')
ax.plot([airfoil[bordo_de_ataque,0],airfoil[bordo_de_fuga,0]],[airfoil[bordo_de_ataque,1],airfoil[bordo_de_fuga,1]],'r-')
```

```python
airf = np.round(airfoil,4)
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.plot(airf[:,0],airf[:,1],'.')
ax.set_aspect('equal')
ax.plot([airf[bordo_de_ataque,0],airf[bordo_de_fuga,0]],[airf[bordo_de_ataque,1],airf[bordo_de_fuga,1]],'r-')
```
