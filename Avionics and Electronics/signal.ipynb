```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
prop = np.arange(-3.4,40,0.01) % 10 < 4
read = np.arange(-3.4,40,0.01) % 1 < 0.3
```

```python
fig = plt.figure(figsize=(15,5))
prop_ax = fig.add_subplot(2,1,1)
prop_ax.axis('off')
prop_ax.plot(prop)
read_ax = fig.add_subplot(2,1,2)
read_ax.axis('off')
read_ax.plot(read)
```
